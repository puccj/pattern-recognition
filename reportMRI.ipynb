```python
#list all the path of mha images
directory_path = '10159290/images/'
mha_list = [directory_path + file for file in os.listdir(directory_path) if file.endswith('.mha')]
```

```python
#convert all images in nii
for image_path in mha_list:
    image = sitk.ReadImage(image_path)
    name = image_path.split("/")[-1].split(".")[0]  #only keep name without extension
    sitk.WriteImage(image, "output/images/" + name + ".nii")
```

```python
#creates all the images

images = []
for file in mha_list:
    images.append(sitk.ReadImage(file))
```

```python
#get useful info on images

sizes = [image.GetSize() for image in images]
widths = [image.GetWidth() for image in images]
heights = [image.GetHeight() for image in images]
npixels = [image.GetNumberOfPixels() for image in images]
spacing = [image.GetSpacing() for image in images]
```

```python
#graph those info

from collections import Counter
import matplotlib.pyplot as plt

# Use Counter to count occurrences
occurrences = Counter(sorted(sizes))

# Extract data for plotting
values = [f"{x}" for x in occurrences.keys()]
counts = list(occurrences.values())

values_f = []
counts_f = []

#remove all the 1 values
for i in range(len(counts)):
    if counts[i] > 2:
        counts_f.append(counts[i])
        values_f.append(values[i])

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Plot the bar chart
plt.bar(values_f, counts_f, color='blue')

# Rotate X-axis labels for better readability
plt.xticks(rotation=90, ha='center')

#plt.xlim(-0.5, len(values_f) -0.5)

# Add labels and title
plt.xlabel('Sizes')
plt.ylabel('Occurrences')
plt.title('Most common sizes of the images')

# Show the plot
plt.show()
```
