```python
import os
import SimpleITK as sitk
```

```python
path = 'data/yesno_backup'
prefix = 'yes_'

yes_name = []
yes_buffers = []
yes_dimensions = []
yes_directions = []
yes_keys = []
yes_metadata = []
yes_components = []
yes_numbers = []
yes_origins = []
yes_types = []
yes_values = []
yes_sizes = []
yes_size_component = []
yes_spacing = []

for root, dirs, files in os.walk(path):
    for file in files:
        if file.startswith(prefix):
            image = sitk.ReadImage(root + '/' + file)
            yes_name.append(file)
            try:
                yes_buffers.append(image.GetBufferAsFloat())
            except:
                print('no float')
                yes_buffers.append(image.GetBufferAsInt16())
            yes_dimensions.append(image.GetDimension())
            yes_directions.append(image.GetDirection())
            yes_keys.append(image.GetMetaDataKeys())
            yes_metadata.append(image.GetMetaData(key) for key in yes_keys)
            yes_components.append(image.GetNumberOfComponentsPerPixel())
            yes_numbers.append(image.GetNumberOfPixels())
            yes_origins.append(image.GetOrigin())
            yes_types.append(image.GetPixelIDTypeAsString())
            yes_values.append(image.GetPixelIDValue())
            yes_sizes.append(image.GetSize())
            yes_size_component.append(image.GetSizeOfPixelComponent())
            yes_spacing.append(image.GetSpacing())
```
