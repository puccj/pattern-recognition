```python
path = 'moose_test1'
prefix = 'no_'

buffers = []
dimensions = []
directions = []         #####
keys = []
metadata = []           
components = []
numbers = []
origins = []
types = []
values = []
sizes = []
size_component = []
spacing = []            #####

for root, dirs, files in os.walk(path):
    for file in files:
        if file.startswith(prefix):
            image = sitk.ReadImage(root + '/' + file)
            try:
                buffers.append(image.GetBufferAsFloat())
            except:
                print('no float')
                buffers.append(image.GetBufferAsInt16())
            dimensions.append(image.GetDimension())
            directions.append(image.GetDirection())
            keys.append(image.GetMetaDataKeys())
            metadata.append(image.GetMetaData(key) for key in keys)
            components.append(image.GetNumberOfComponentsPerPixel())
            numbers.append(image.GetNumberOfPixels())
            origins.append(image.GetOrigin())
            types.append(image.GetPixelIDTypeAsString())
            values.append(image.GetPixelIDValue())
            sizes.append(image.GetSize())
            size_component.append(image.GetSizeOfPixelComponent())
            spacing.append(image.GetSpacing())
```
